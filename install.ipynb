```
from imageai.Detection.Custom import CustomObjectDetection
import cv2
import matplotlib.pyplot as plt

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()

# choose the best model
detector.setModelPath("/content/drive/My Drive/test/model yolo/models/detection_model-ex-013--loss-0002.522.h5") 
detector.setJsonPath("/content/drive/My Drive/test/model yolo/json/detection_config.json")
detector.loadModel()

in_image = ['/content/drive/My Drive/test/validation/images/10 014.jpg', '/content/drive/My Drive/test/validation/images/11 107.jpg', '/content/drive/My Drive/test/validation/images/115.jpg','/content/drive/My Drive/test/validation/images/15 035.jpg','/content/drive/My Drive/test/validation/images/16 206.jpg','/content/drive/My Drive/test/validation/images/17 057.jpg','/content/drive/My Drive/test/validation/images/18 111.jpg','/content/drive/My Drive/test/validation/images/19 147.jpg','/content/drive/My Drive/test/validation/images/2 087.jpg','/content/drive/My Drive/test/validation/images/20 053.jpg','/content/drive/My Drive/test/validation/images/3 055.jpg','/content/drive/My Drive/test/validation/images/4 042.jpg','/content/drive/My Drive/test/validation/images/5 021.jpg','/content/drive/My Drive/test/validation/images/6 042.jpg','/content/drive/My Drive/test/validation/images/7 049.jpg','/content/drive/My Drive/test/validation/images/8 142.jpg','/content/drive/My Drive/test/validation/images/9 195.jpg','/content/drive/My Drive/test/validation/images/40.jpg','/content/drive/My Drive/test/validation/images/2 198.jpg','/content/drive/My Drive/test/validation/images/10 255.jpg']
out_image = ['/content/drive/My Drive/test/output/YOLOv3/1.jpg','/content/drive/My Drive/test/output/YOLOv3/2.jpg', '/content/drive/My Drive/test/output/YOLOv3/3.jpg','/content/drive/My Drive/test/output/YOLOv3/4.jpg','/content/drive/My Drive/test/output/YOLOv3/5.jpg','/content/drive/My Drive/test/output/YOLOv3/6.jpg','/content/drive/My Drive/test/output/YOLOv3/7.jpg','/content/drive/My Drive/test/output/YOLOv3/8.jpg','/content/drive/My Drive/test/output/YOLOv3/9.jpg','/content/drive/My Drive/test/output/YOLOv3/10.jpg','/content/drive/My Drive/test/output/YOLOv3/11.jpg','/content/drive/My Drive/test/output/YOLOv3/12.jpg','/content/drive/My Drive/test/output/YOLOv3/13.jpg','/content/drive/My Drive/test/output/YOLOv3/14.jpg','/content/drive/My Drive/test/output/YOLOv3/15.jpg','/content/drive/My Drive/test/output/YOLOv3/16.jpg','/content/drive/My Drive/test/output/YOLOv3/17.jpg','/content/drive/My Drive/test/output/YOLOv3/18.jpg','/content/drive/My Drive/test/output/YOLOv3/19.jpg','/content/drive/My Drive/test/output/YOLOv3/20.jpg']

for i in range(0, len(in_image)):
  detections = detector.detectObjectsFromImage(input_image=in_image[i], output_image_path=out_image[i])
  for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
    
  # show detected image
  img = cv2.imread(out_image[i])
  plt.imshow(img)
  plt.show()
```
